```python
import pandas as pd
import numpy as np
```

```python
# Importar a base dados
from google.colab import files
uploaded = files.upload()
```

```python
# Ler o arquivo csv
df = pd.read_csv('Quantum_dot_data.csv')
```

```python
# Obter dados do arquivo
print("Shape:", df.shape)
print("Primeiras linhas:")
print(df.head())
```

```python
# Obter nomes das colunas
print("Nomes das colunas:")
print(df.columns)
```

```python
# Analisar a coluna Compound (target/qualitativa)
print(df["Compound"].value_counts())
```

```python
# Checar por valores faltantes
print(df.isnull().sum())
```

```python
# Preenchendo dados faltantes com a média da coluna
numeric_columns = df.select_dtypes(include=[np.number]).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())
```

```python
# Separar a coluna Compound (y, target) das demais colunas (x, preditores)
X = df.drop("Compound", axis=1)
y = df["Compound"]
```

```python
# Transformar a variável categórica y em numérica (classificação binária)
y_binario = pd.get_dummies(y, drop_first=True).astype(int)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Dividir os dados entre conjunto de dados de treino e conjunto de dados de teste
X_train, X_test, y_train, y_test = train_test_split(X, y_binario, test_size=0.2, random_state=42, stratify = y_binario)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Estantandardizar (padronizar, normalizar) as features (preditores, x)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Treinar o modelo de regressão logística
rl_model = LogisticRegression(max_iter=20, random_state=42)
rl_model.fit(X_train_scaled, y_train)
```

```python
# Fazer predições nos dados de teste
y_pred = rl_model.predict(X_test_scaled)
```

```python
# Fazer predições sobre a probabilidade de uma observação pertencer a uma classe
y_pred_proba = rl_model.predict_proba(X_test_scaled)
```

```python
# Calcular as métricas para regressão logística (acurácia, sensibilidade, especificidade)
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Calcular acurácia, sensibilidade, espeficidade
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
# Gerar a matriz de confusão
from sklearn.metrics import confusion_matrix
```

```python
# Gerar e mostrar a matriz de confusão
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# Mostrar as métricas calculadas
print("Acurácia:", accuracy)
print("Precisão:", precision)
print("Sensibilidade:", recall)
print("F1-score:", f1)
```

```python
# Importar pacote de Análise Discriminante Linear
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
# Treinar modelo de Análise Discriminante Linear
lda_model = LinearDiscriminantAnalysis()
lda_model.fit(X_train_scaled, y_train)
```

```python
# Fazer as predições usando LDA
y_pred_lda = lda_model.predict(X_test_scaled)
```

```python
# Calcular as métricas de performance do modelo LDA (acurácia, sensibilidade, especificidade)
accuracy_lda = accuracy_score(y_test, y_pred_lda)
precision_lda = precision_score(y_test, y_pred_lda)
recall_lda = recall_score(y_test, y_pred_lda)
```

```python
# Mostrar os valores calculados
print("Acurácia (LDA):", accuracy_lda)
print("Precisão (LDA):", precision_lda)
print("Sensibilidade (LDA):", recall_lda)
```

```python
# Calcular a matriz de confusão para o modelo LDA
cm_lda = confusion_matrix(y_test, y_pred_lda)
print("Matriz de Confusão (LDA):")
print(cm_lda)
```
