```python
# importar bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
# Load the data
from google.colab import files
uploaded = files.upload()
```

```python
# Read csv file
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)
```

```python
# Estandardizar os dados
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
# Aplicar o método PCA
pca = PCA()
X_pca = pca.fit_transform(df_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
# Criar dataframe dos resultados do PCA
df_pca = pd.DataFrame(X_pca)
df_pca
```

```python
# Visualizar scree plot
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

```python
# Visualizar scree plot dos 10 primeiros componentes
plt.plot(np.cumsum(pca.explained_variance_ratio_[:10]))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

```python
# Visualizar plot PCA PC1 versus PC2
plt.scatter(df_pca[0], df_pca[1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
